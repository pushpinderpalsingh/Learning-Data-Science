```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.linear_model import LogisticRegression
```

```python
X,y = make_blobs(centers=4,random_state=10)
```

```python
X.shape, y.shape
```

```python
plt.scatter(X[:,0],X[:,1], c=y)
```

```python
model = LogisticRegression()
```

```python
model.fit(X,y)
```

```python
model.classes_
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_col)
for index,[[m1,m2],c] in enumerate(zip(model.coef_,model.intercept_)):
    x1 = np.linspace(-10,10,100)
    x2 = (-m1/m2)*x1 -(c/m2)
    plt.plot(x1,x2)
```

```python
colors = ["Red", "Yellow", "Green", "Orange", "Black"]
```

```python
y_col = [colors[item] for item in y]
```

```python
model.score(X,y)
```

```python
model.predict_proba(X).argmax(axis=1)
```
