```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
```

```python
X,y = make_regression(n_features=5,noise=20,random_state=1,bias=100)
```

```python
np.shape(X) 
```

```python
plt.scatter(X[:,4],y)
plt.scatter(X[:,3],y)
plt.scatter(X[:,2],y)
plt.scatter(X[:,1],y)
plt.scatter(X[:,0],y)
```

```python
plt.scatter(X,y)
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
yp = model.predict(X)
```

```python
plt.scatter(X,y)
plt.scatter(X,yp)
```
