```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m,c = 4, 1000
X = np.linspace(10,100,100)
noise = np.random.randn(100)*3000
y = X*m *X + c + noise
X_mod = (X-X.mean())/X.std()
y_mod = (y-y.mean())/y.std()
X,y = X_mod,y_mod
plt.scatter(X,y)
```

```python
class LinearReg:
    def __init__(self,lr = 0.01,iters = 10):
        self.lr = lr
        self.iters = iters
    
    def fit(self,X,y):
        self.a_ = np.random.random()
        self.b_ = np.random.random()
        self.c_ = np.random.random()
        errors = []
        for i in range(self.iters):
            errors.append(self.gradient_desc(X,y))
        return errors
    
    def gradient_desc(self,X,y):
        delta_a,delta_b,delta_c,error = self.gradient(X,y)
        self.a_ += delta_a
        self.b_ += delta_b
        self.c_ += delta_c
        return error
    
    def gradient(self,X,y):
        yh = self.hypo(X)
        mse = np.mean((yh - y)**2)
        dmseda = np.mean(2*(yh-y)*X*X)
        dmsedb = np.mean(2*(yh-y)*X)
        dmsedc = np.mean(2*(yh-y))
        
        return -dmseda * self.lr,-dmsedb * self.lr,-dmsedc * self.lr,mse
    
    def hypo(self,X):
        return X * self.a_ *X + X*self.b_ + self.c_    
    
```

```python
model = LinearReg(lr=0.3,iters=10)

```

```python
error = model.fit(X,y)
plt.plot(error)
# error
```

```python
yh = model.hypo(X)
plt.scatter(X,y)
plt.scatter(X,yh)
```
