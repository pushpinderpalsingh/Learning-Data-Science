```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,100)
```

```python
y = x
```

```python
plt.figure()
plt.plot(x,y**2,label = "square")
plt.plot(x,y**3,label = "cube")
plt.legend()
plt.ylim(-100,100)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
x = np.linspace(-np.pi,np.pi,100)
```

```python
plt.figure()
plt.plot(x,np.sin(x), label = "Sin")
plt.plot(x,np.cos(x),label = "Cos")
plt.plot(x,np.tan(x),label = "Tan")
plt.legend()
plt.ylim(-4,4)
plt.xlabel("X-axis")
plt.show()
```

```python
plt.figure(figsize=(10,10))
theta = np.linspace(0,100*np.pi,1500)
rad = np.linspace(0,100,1500)
plt.plot(rad*np.cos(theta),rad*np.sin(theta))
```
