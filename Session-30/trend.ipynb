```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
data = pd.read_csv("AAPL.csv")
```

```python
data.head()
```

```python
data_high = data.High
```

```python
plt.plot(data_high)
```

```python
window = 30
half = window // 2
out = []
for i in range(len(data_high)):
    if i < half:
        cut = data_high[:i+half]
    else:
        cut = data_high[i-half:i+half]
    out.append(np.mean(cut))
```

```python
plt.plot(out)
plt.plot(data_high)
plt.plot(data_high-out)
```
