```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m,c = 4,0
```

```python
X = np.linspace(10,50,100)
noise = np.random.randn(100) * 20
```

```python
y = m * X  + c + noise
plt.scatter(X,y)
```

```python
class CustomLinearReg:
    def __init__(self,lr=0.1,iters=10):
        self.lr = lr
        self.iters = iters
    
    def fit(self,X,y):
        self.coef_ = np.random.random()
        errors = []
        for i in range(self.iters):
            error = self.gradientdesc(X,y)
            errors.append(error)
        return errors
    
    def gradientdesc(self,X,y):
        deltaM,mse = self.gradient(X,y)
        self.coef_ += deltaM
        return mse
    
    def gradient(self,X,y):
        yh = self.hypo(X)
        mse = np.mean((yh-y)**2)
        dmse = np.mean(2*X*(yh-y))
        return -dmse*self.lr,mse
        
    def hypo(self,X):
        return X*self.coef_
```

```python
model = CustomLinearReg(lr=0.0005,iters=4)
```

```python
errors = model.fit(X,y)
plt.plot(errors)
```

```python
yh = model.hypo(X)
errors
```
