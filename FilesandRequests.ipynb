```python
import requests
import random
from PIL import Image
```

```python
request = requests.get("https://www.googley.com/")
while request.status_code != 200:
    imgURL = r"http://graph.facebook.com/{}/picture?type=square".format(random.randint(10000000000,99999999999))

    request = requests.get(imgURL)
```
