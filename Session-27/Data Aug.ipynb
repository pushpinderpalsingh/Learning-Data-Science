```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img
```

```python
img = np.array(load_img("../datasets/panda.jpg"))
```

```python
plt.imshow(img)
```

```python
gen = ImageDataGenerator(rotation_range=25,horizontal_flip=True,shear_range=.2,zoom_range=.4)
```

```python
flow = gen.flow(np.array([img]))
```

```python
batch = next(flow)
```

```python
batch
```

```python
batch = next(flow)
for i in range(len(batch)):
    plt.figure()
    plt.imshow(batch[i].astype(np.uint8))
    plt.show()
```
