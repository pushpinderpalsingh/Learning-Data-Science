```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img
```

```python
img = np.array(load_img("../datasets/RGB_Matte.png", target_size=(200, 200)))
```

```python
img.shape
```

```python
img_grey = img.mean(axis=2)
```

```python
plt.imshow(img_grey,cmap="gray")
```

```python
fil = np.array([[1,0,-1],
                [1,0,-1],
                [1,0,-1]])
```

```python
fil
```

```python
img_grey.shape
```

```python
padded_img = np.pad(img_grey,1)

```

```python
padded_img.shape
```

```python
result = np.empty_like(img_grey)
```

```python
result.shape
```

```python
for row in range(result.shape[0]):
    for col in range(result.shape[1]):
        cut = padded_img[row:row+3,col:col+3]
        result[row,col] = np.absolute(np.mean(fil*cut))
```

```python
plt.imshow(result,cmap="gray")
```
